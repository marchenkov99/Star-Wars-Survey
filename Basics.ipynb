## Reading the file

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
print(len(star_wars))
star_wars.head(1)
```

### Columns

```python
star_wars.columns
```

### Remove `RespondentID` NaN columns.

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
len(star_wars)
```

```python
star_wars.head()
```

### Convert `Yes/No` to `True/False` type

```python
yes_no = {
    "Yes": True,
    "No": False
}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].unique()
```

### Convert columns[3:6] to `True` and `False`

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

### Renaming `columns[3:6]`

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})

star_wars.head()
```

### Convering rank columns

```python
star_wars.iloc[:, 9:15] = star_wars.iloc[:, 9:15].astype(float)

star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
})

star_wars.head(2)
```

### Mean values

```python
mean_ranking = star_wars.iloc[:, 9:15].mean()
mean_ranking
```

### Bar chart of each ranking

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(len(mean_ranking)) + 0.75
width = 0.5

print(star_wars.describe(include=['O']))
plt.bar(x, mean_ranking, width, color='c')
plt.tick_params(bottom='off', top='off')
```

### Number of views

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

plt.bar(np.arange(6) + 0.75, males[males.columns[9:15]].sum(), 0.25, color=(0.2588,0.4433,1.0))
plt.bar(np.arange(6) + 1, females[females.columns[9:15]].sum(), 0.25, color=(1.0,0.5,0.62))
plt.tick_params(bottom='off', top='off')
plt.legend(("Male", "Female"))
```
